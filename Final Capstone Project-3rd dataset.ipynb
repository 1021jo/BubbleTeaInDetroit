```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

```

3rd Dataset: University Locations in Detroit Area

```python
url = "https://www.collegesimply.com/colleges-near/michigan/detroit"
page=requests.get(url)
html_string = page.text
soup = BeautifulSoup(html_string, 'lxml') # Parse the HTML as a string
table = soup.find_all('table')[0] # Grab the first table
    
df = pd.read_html(str(table))
df=df[0]
```

```python
df
```

```python
df=df.loc[:,['College','Location']]
college = df["College"].str.split(":", n = 1, expand = True)
Enrollment=college.iloc[:,1].str.split("Tuition", n = 1, expand = True)
Enrollment
df['College Name']=college.iloc[:,0]
df['Enrollment']=Enrollment.iloc[:,0]
collegedf=df[['College Name','Enrollment','Location']]
collegedf
```

Add coordinates to the college dataframe

```python
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
collegedf['Latitude']=''
collegedf['Longitude']=''
final=collegedf
final
```

```python
i=0
for address in collegedf.iloc[:,2]:
    geolocator = Nominatim()
    location = geolocator.geocode(address)
    latitude = location.latitude
    longitude = location.longitude
    final.iloc[i,3] = latitude
    final.iloc[i,4] = longitude
    i=i+1
```

```python
final
```

```python
final['College Name']= final['College Name'].str.replace('Enrollment','')

final = final.loc[final['Location']!="Ann Arbor, Michigan"]
final = final.loc[final['Location']!="Ypsilanti, Michigan"]
```

```python
# Searched online to add the enrollment for missing data

final.iloc[10,1]=1103
final.iloc[14,1]=1318
final.iloc[18,1]=958
final=final.drop([11, 16])
final = final.reset_index(drop=True)
final
```
